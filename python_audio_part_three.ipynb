# Handling audio data in Python part three

## Time domain analysis

Now we are going to work with some urban audio recordings, which are far more complex than the signals we have previously dealt with

First of all get all of our imports and setup calls:

```python
%matplotlib inline
import numpy as np
from scipy.io import wavfile
import scipy.stats as sps
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

pylab.rcParams['figure.figsize'] = 16, 8
pylab.rcParams['font.size'] = 16
```

Read in the audio data from a mono recording made on a Brooklyn street

```python
fs, samples = wavfile.read('audio/brooklyn_street.wav')
```

Lets check out some of the recordings details

```python
dur = float(len(samples)) / fs
sample_len = len(samples)

print 'Sample rate: ', fs, 'Hz'
print 'Length in samples: ', sample_len
print 'Length in seconds: ', '%.3f' % dur
print 'Sample data type: ', samples.dtype
```

The sample data is in the form of 16 bit integers so lets convert to floats

```python
samples = samples / 32768.0
```

Now we want to get a feel for the urban recording so lets plot it in its entirety in the time domain with the time in seconds along the x axis

First we create our time value array for plotting time across the x axis

```python
T = np.linspace(0, dur, num=sample_len)
```

Now we plot our urban soundscape in the time domain

```python
plt.plot(T, samples)
plt.ylim([-1.1, 1.1])
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

This gives us an idea about the **"amplitude envelope"** of our recording. This means we can determine how the volume changes over time

To get a better look at the amplitude envelope we take the absolute of our audio samples, meaning we change the signs of all negative values to positive

```python
ab_samples = np.abs(samples)
```

Plot the absolute waveform to see how it looks, limiting the y-axis range to the length of data

```python
plt.plot(T, ab_samples)
plt.ylim([0, 1.1])
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street absolute amplitude')
plt.xlabel('Time (s)')
plt.ylabel('Absolute amplitude')
```

The raw absolute amplitude data gives us some clues about the evolution of the soundscape over time

The amplitude fluctuations are very fast and make it difficult to pick out individual sonic events

In order to smooth out these very short time fluctuations it helps to employ a moving average filter. This involves running a **window** across the data and calculating an average or other statistical value for each of these windows

Now we need to determine some characteristics of our windowing function

We are going to create a moving average window first so we need to define our window size and the hop size that this window will take across our signal. We will start with an arbritrary width window, which jumps along a half of the window width at a time, resulting in a 50% window overlap

```python
window_size = 8192
hop_size = window_size/2
```

Lets check how long our window is in milliseconds

```python
print 'Window length in ms: ', '%.2f' % (float(window_size)/float(fs)*1000)
```

Window size can be adjusted to suit the signal you are analyzing. Longer window sizes mean you lose temporal resolution, but it can help to smooth out excessively impulsive signals

The most basic form of window is the **rectangular window** shown below to the left. This type **does not apply** any kind of function to the windows sample points.

The problem with a rectangular window is that its sharp leading and trailing edges can actually introduce artifacts to the analysis of that window, especially when working in the frequency domain

A solution to this is to multiply the sample window by a function, which acts to gradually fade the windowed samples in and out at the leading and trainling edges. This helps to reduce these artifacts.

The centre window function below is called a **triangular/Bartlett** window and the right one is called a **Hanning window**. They both act to reduce these artifacts when analysing each window of sample data

<img src="http://www.mikroe.com/img/publication/various-electronic-books/digital-filter-design/chapter/ch2/fig2-3-2.gif"  style="float: left;" width=33%>
<img src="http://www.mikroe.com/img/publication/various-electronic-books/digital-filter-design/chapter/ch2/fig2-3-6.gif"  style="float: right;" width=30%>
<img src="http://www.mikroe.com/img/publication/various-electronic-books/digital-filter-design/chapter/ch2/fig2-3-4.gif" width=30%>
    

In the following examples we will use a Triangle/Bartlett window function for each of our sample windows

Lets create our window function using Numpy's built in functions and plot it to see how it looks

```python
w = np.bartlett(window_size)

plt.plot(w)
plt.ylim([0, 1.1])
plt.xlim([0, window_size])
plt.title('Triangular window')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
```

As you can see, we have created a function that has a max value of 1.0 and min value of 0.0, which we can multiply the samples of our sample window by

Now we can take a sample window from our absolute amplitude audio signal and multiply it by our window function and plot the result. In the code below I've taken a sample window from 10 seconds into our urban audio recording and applied our triangular window function

```python
plt.plot(w * ab_samples[fs*10:(fs*10)+window_size])
plt.xlim([0, window_size])
```

You can see how the windowing function is fading the absolute samples in and out in amplitude 

Now we can run our window over the entire audio signal length and create a new array of filtered values. In this case we are taking the arithmetic mean of each of our windowed samples

```python
window_avg = [np.mean(w*ab_samples[i:i+window_size]) for i in range(0, len(ab_samples), hop_size)if i+window_size <= len(ab_samples)]
```

Note that we've filtered our original sample values by averaging windowed portions of it so the length of our resultant array is much shorter

```python
print 'Length of audio samples array:    ', len(ab_samples)
print 'Length of windowed average array: ', len(window_avg)
```

Create a new time array for plotting using our new windowed average array length

```python
T = np.linspace(0, dur, num=len(window_avg))
```

Now we can plot our windowed mean amplitude signal. You'll notice that its smoothed out a lot of the jagged spikes and allows us to see more of the distinct peaks in amplitude over time

```python
plt.plot(T, window_avg)
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street')
plt.xlabel('Time (s)')
plt.ylabel('Absolute amplitude')
```

The question then is: is this enough information to identify individual sound events within an urban audio recording?

It's definately a step in the right direction as we can clearly identify distinct peaks in amplitude that must correspond to some kind of event. However, quieter, but still salient sound events may have an amplitude below the background noise level, which means they will not exhibit visible amplitude spikes in the time domain

In order to delve deeper into the signal we need to look from a different angle. This means moving into the frequency domain

## Frequency domain analysis

In addition to working with statistical data derived from each window in the time domain, the analysis of each windows frequency data can provide much more information about how the signals spectral characteristics change over time

Lets start by taking the Fourier Transform (FT) of the first window of our original audio signal, making sure we first multiply the window samples by our windowing function

```python
dft_output = np.fft.rfft(w * samples[0:window_size])
```

Dont forget that the output of our FT is an array of frequency data, so for each of our sample windows we have another array of frequency data

Now we calculate the magnitude spectrum of our windows FT

```python
magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(dft_output) for i in dft_output]
```

Create our frequency array for plotting on our y-axis

```python
freqs = np.linspace(0,fs/2, num=len(dft_output))
```

Plotting our magnitude spectrum (MS) data gives us the frequency content of our first window

```python
plt.plot(freqs, magnitude_spectrum, 'g')
plt.xlim([20, 20000])
plt.title('Magnitude spectrum of first window of urban audio')
plt.xlabel('Frequency (linear Hz)')
plt.ylabel('Amplitude')
```

The problem here is that the majority of our frequency content/energy is crammed up in the lower frequency range

Urban audio recordings will tend to be like this as most of the acoustic energy is present in the lower frequency range. Plotting the FT using a linear frequency scale is generally inadequate to vizualize urban audio data, so we need to display the frequency data on the x-axis on a **logarithmic scale**

To help us visualize the change in frequency content over time in a real world signal, we will plot the MS of the first 3 windows of our signal using a logarithmic frequency scale on the x-axis

```python
dft_output = np.fft.rfft(w * samples[0:window_size])

magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(dft_output) for i in dft_output]

plt.subplot(3, 1, 1)
plt.semilogx(freqs, magnitude_spectrum, 'g')
plt.xlim([20, 20000])
plt.ylim([0, 0.01])
plt.ylabel('Amplitude')
plt.title('Magnitude spectrum of first, second and third windows of urban audio')
plt.grid(b=True, which='both', color='0.6',linestyle='--')
plt.text(2000, 0.005, 'STD = %.5f' % np.std(magnitude_spectrum))

dft_output = np.fft.rfft(w * samples[hop_size:window_size+hop_size])

magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(dft_output) for i in dft_output]

plt.subplot(3, 1, 2)
plt.semilogx(freqs, magnitude_spectrum, 'g')
plt.xlim([20, 20000])
plt.ylim([0, 0.01])
plt.ylabel('Amplitude')
plt.grid(b=True, which='both', color='0.6',linestyle='--')
plt.text(2000, 0.005, 'STD = %.5f' % np.std(magnitude_spectrum))

dft_output = np.fft.rfft(w * samples[hop_size * 2:window_size + hop_size * 2])

magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(dft_output) for i in dft_output]

plt.subplot(3, 1, 3)
plt.semilogx(freqs, magnitude_spectrum, 'g')
plt.xlim([20, 20000])
plt.ylim([0, 0.01])
plt.ylabel('Amplitude')
plt.xlabel('Frequency (logarithmic Hz)')
plt.grid(b=True, which='both', color='0.6',linestyle='--')
plt.text(2000, 0.005, 'STD = %.5f' % np.std(magnitude_spectrum))
```

You can see how the MS changes over time, with new spectral peaks developing over a very short time period. Don't forget that the difference in time between these plots is:

```python
print 'Difference in time between windows: ', '%.2f' % (1000 * float(hop_size)/float(fs)), 'ms'
```

We now need to hop along our original audio samples FT'ing each window and storing the results in an array of arrays

```python
mag_spec_array_complex = [np.fft.rfft(w * samples[i:i + window_size]) for i in range(0, len(samples), hop_size)if i + window_size <= len(samples)]
```

We then need to iterate through this array of arrays and convert the containing complex number arrays to magnitude spectrum arrays

```python
mag_spec_array = []
for x in range(0, len(mag_spec_array_complex)):
    mag_spec_array.append([np.sqrt(i.real**2 + i.imag**2)/len(mag_spec_array_complex[x]) for i in mag_spec_array_complex[x]])
```

Now that we have our array of MS arrays, we can begin to compute single value statistics for each MS array over time

The first spectral feature value we will compute is the **spectral centroid** of each window's MS. The spectral centroid can be considered the "mean" frequency of the MS and has perceptual links to a sounds "brightness"

```python
spec_cent_array = np.zeros(len(mag_spec_array))

for x in range(0, len(mag_spec_array)):
    spec_cent_array[x] = np.sum(mag_spec_array[x]*freqs) / np.sum(mag_spec_array[x])
```

When we plot the result of this computation, we are plotting the spectral "mean" or "centre of gravity" over time

```python
plt.plot(T,spec_cent_array, 'r')
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street - spectral centroid over time')
plt.xlabel('Time (s)')
plt.ylabel('Spectral centroid (Hz)')
```

It will help to compare this new representation of the urban audio signal to its absolute amplitude values for comparison

```python
fig, ax1 = plt.subplots()

ax1.plot(T, window_avg)

for tl in ax1.get_yticklabels():
    tl.set_color('b')

ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Absolute amplitude')

ax2 = ax1.twinx()

ax2.plot(T,spec_cent_array, 'r', alpha=0.7)
ax2.set_ylabel('Spectral centroid (Hz)')

for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.xlim([0, np.max(T)])
```

We can see that the spectral centroid does not always follow the signals amplitude

This suggests that it can be used to identify sound sources that are low in amplitude, but are different in frequency to the background noise of the urban audio recording, ie: it is providing us with more information about the audio signal

A common technique that is widely used in identifying changes in signal state is looking at the **differences between a windows spectral characteristics over time**

In the plot below I've taken the differences between adjacent windows spectral centroid and plotted them over time

```python
plt.plot(T[0:len(np.diff(spec_cent_array))], np.diff(spec_cent_array), 'r')
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street - spectral centroid differences over time')
plt.xlabel('Time (s)')
plt.ylabel('Spectral centroid differences (Hz)')
```

You can see distinct peaks on the plot, which signify quick changes in spectral energy that are generally caused by sources that begin and end abruptly

Another spectral feature that helps in analysis is the **spectral flatness** of a windows MS.

A window with a flatness value that tends towards 1.0 indicates that its MS is more like that of white noise (equal energy across all frequency bands), whereas if it tends towards 0.0 its frequency energy is concerntrated in a small number of bands and the sound is more tonal in nature

The spectral is computed by taking the ratio of the MS's geometric mean to its arithmetic mean

```python
spec_flat_array = np.zeros(len(mag_spec_array))

for x in range(0, len(mag_spec_array)):
    spec_flat_array[x] = sps.mstats.gmean(mag_spec_array[x]) / np.mean(mag_spec_array[x]) 
```

When plotted you can see that it reveals more detail of the signals spectral evolution over time

```python
plt.plot(T,spec_flat_array, 'r')

plt.xlim([0, np.max(T)])
plt.title('Brooklyn street - spectral flatness over time')
plt.xlabel('Time (s)')
plt.ylabel('Spectral flatness')
```

Now we can also plot this alongside the signals absolute amplitude for comparison

```python
fig, ax1 = plt.subplots()

ax1.plot(T, window_avg)

for tl in ax1.get_yticklabels():
    tl.set_color('b')

ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Absolute amplitude')

ax2 = ax1.twinx()

ax2.plot(T,spec_flat_array, 'r', alpha=0.7)
ax2.set_ylabel('Spectral flatness')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.xlim([0, np.max(T)])
```

The peaks and troughs have their own pattern, so may reveal more about the signals content over time

Once again the spectral flatness differences between window MS's can give us another clue to a signals changes over time in the frequency domain

```python
plt.plot(T[0:len(np.diff(spec_flat_array))], np.diff(spec_flat_array), 'r')
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street - spectral flatness differences over time')
plt.xlabel('Time (s)')
plt.ylabel('Spectral flatness')
```

Another metric that can be extracted from our windowed MS is the **spectral flux**. This describes how quickly the spectral energy is changing between windowed MS. It is used to determine the beginning/onset of notes in musical pieces.

It is computed by calculating the euclidean distance between adjacent MS's in time 

An increase in the spectral flux indicates the onset/beginning of an event as the spectral energy is changing

We need to import the distance library from SciPy's spatial package in order to use the euclidean distance function

```python
from scipy.spatial import distance
```

Then we calculate the distance between adjacent spectral windows and store them in a new array

```python
spec_flux_array = np.zeros(len(mag_spec_array) - 1)

for x in range(0, len(mag_spec_array) - 1):
    spec_flux_array[x] = distance.euclidean(mag_spec_array[x], mag_spec_array[x+1])
```

And plot the resultant data

```python
plt.plot(T[0:len(spec_flux_array)],spec_flux_array, 'r')
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street - spectral flux over time')
plt.xlabel('Time (s)')
plt.ylabel('Spectral flux')
```

```python
fig, ax1 = plt.subplots()

ax1.plot(T, window_avg)

for tl in ax1.get_yticklabels():
    tl.set_color('b')

ax1.set_xlabel('Time (s)')

ax2 = ax1.twinx()

ax2.plot(T[0:len(spec_flux_array)],spec_flux_array, 'r', alpha=0.7)

for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.xlim([0, np.max(T)])
```

## Visualizing in the time and frequency domain

The visualization of complex waveforms both in the time and frequency domain can help in understanding their underlying characteristics

The **spectrogram** is the visual representation of a time domain signals frequency content as they vary with time. The dimensions of this representation are time on the x-axis, frequency on the y-axis, with the "z-axis" representing amplitude in the form of pixel color

We have actually already created the data to generate a spectrogram. The array of magnitude spectrum arrays for each window contains all the data we need

```python
spec_array = np.array(mag_spec_array).T
plt.imshow(spec_array, aspect='auto', origin='lower')
plt.set_cmap('gnuplot')
```

The problem with using this data as it is, is that the range it is in does not relate well to color, resulting in a very faint spectral image

To help us understand how the spectrogram works, we will generate something called a **chirp** or **swept sine** signal. This is a signal that starts at a low frequency and rises in frequency over time

```python
from scipy.signal import chirp

dur = 3

chir_samp_len = fs * dur

chirp_win_size = 4096

T_chirp = np.linspace(0, dur, chir_samp_len)

chirp_samps = chirp(T_chirp, f0=20, t1=dur, f1=20000, method='linear')
```

Matplotlib has a built in spectrogram plotting function that handles the details for us, so we will use this to plot our chirp

Notice how I've set the window size to a different value to the previous analysis work. Try adjusting this value and see what effect this has on the spectrogram.

```python
plt.specgram(chirp_samps, NFFT = chirp_win_size, noverlap = chirp_win_size/2, Fs = fs, mode = 'magnitude', scale = 'dB');
plt.ylim([0, 20000])
plt.xlim([0, np.max(T_chirp)])
plt.set_cmap('gist_rainbow_r')
plt.title('Chirp/linear swept sine - spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid(b=True, which='both', color='0.6',linestyle='--')
```

Increasing the window size gives us more frequency resolution, but impaired time resolution. Decreasing the size impairs our frequency resolution, but improves the time resolution

We have the amount of window overlap set at 50%. Vary this to see its effect. If this is very small, adjacent windows will not overlap as much, so you will get less pixel density along the x-axis in our spectrogram. Increasing the overlap will result in more MS windows being caluculated, so a higher pixel density will be achieved

Now lets generate a spectrogram of our urban recording

```python
plt.specgram(samples, NFFT = window_size, noverlap = window_size/2, Fs = fs, mode = 'magnitude', scale = 'dB');
plt.ylim([0, 17000])
plt.xlim([0, np.max(T)])
plt.set_cmap('gist_rainbow_r')
plt.title('Brooklyn street - spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid(b=True, which='both', color='0.6',linestyle='--')
```
